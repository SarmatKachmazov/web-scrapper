```python
with open('index89.html') as file:
    src = file.read()
soup = bs(src, "lxml")
```

```python
noClearLinks1 = []
for a in soup.find_all('a', class_ = 'cmc-link', href=True):
    #print("Found the URL:", a['href'])
    link = a['href']
    noClearLinks1.append(link)
#print(links)

noClearLinks2 = []
for i in noClearLinks1:
    if "currencies" in i:
        #print(i)
        noClearLinks2.append(i)
    
#print(len(noClearLinks2))

links = []
for k in noClearLinks2:
    if 'markets' in k:
        continue
    if 'period' in k:
        continue
    if 'sitemap' in k:
        continue
    if 'fiat' in k:
        continue
    else:
        print(k)
        links.append(k)


print(len(links))
```

```python
href = soup.find(class_ = 'sc-f7a61dda-2 efhsPu').find_all(class_ = 'cmc-link') # рабочий
readyLinks = {}

for i in href:
    if '$' in i.text:
        continue
    if 'period' in i.get('href'):
        continue
    if 'markets' in i.get('href'):
        continue
    else:
        item_text = i.text
        item_href = i.get("href") 
        item_href = 'https://coinmarketcap.com' + item_href
        print(f'{item_text}: {item_href}')

        readyLinks[item_text] = item_href

with open("readyLinks89", 'w') as file:
    json.dump(readyLinks, file, indent = 4, ensure_ascii = False)
  
```
